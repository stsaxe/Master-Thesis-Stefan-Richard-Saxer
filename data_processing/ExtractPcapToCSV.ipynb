```python
data = pd.read_parquet(dataPathCSV + r"Processed Files\labeled_and_aggregated_data.parquet")
```

```python
from data_processing.LabelLut import LABEL_LUT_PROD


data = data[data['Source'] != '']
data = data.sort_values(["Label", "File", "Source", "Time"]).copy()
data['Label'] = data['Label'].replace(to_replace=LABEL_LUT_PROD)
```

```python
data.info()
```

```python
size_per_label = data.groupby("Label", group_keys=False).size().reset_index().rename(columns={0: 'Count'})
size_per_label = size_per_label[size_per_label['Count'] >= 20_000]

size_per_label
```

```python
data = data.merge(size_per_label, on="Label", how="inner").reset_index(drop=True)
data.drop('Count', axis=1, inplace=True)
data
```

```python
from data_processing.preprocessing_functions import split_train_val_test

train, val, test = split_train_val_test(data, train_frac=0.8, val_frac=0.1, test_frac=0.1, file_col='File', time_col='Time')
```

```python
train = train.sort_values(["Label", "File", "Source", "Time"]).reset_index(drop=True)
val = val.sort_values(["Label", "File", "Source", "Time"]).reset_index(drop=True)
test = test.sort_values(["Label", "File", "Source", "Time"]).reset_index(drop=True)
```

```python
train.to_parquet(dataPathCSV + r"Processed Files\labeled_and_aggregated_data_train.parquet",
                engine="pyarrow",
                compression="snappy",
                index=False)

val.to_parquet(dataPathCSV + r"Processed Files\labeled_and_aggregated_data_validation.parquet",
                engine="pyarrow",
                compression="snappy",
                index=False)


test.to_parquet(dataPathCSV + r"Processed Files\labeled_and_aggregated_data_test.parquet",
                engine="pyarrow",
                compression="snappy",
                index=False)
```
