# Prepare Dataset for Inference

```python
import os
from pathlib import Path

os.chdir(Path.cwd().parents[0])
print("Now in:", Path.cwd())

dataPathInference = str(Path.cwd()) + r"\data\pcap" + r"\Inference\\"
dataPathProcessedPCAP = str(Path.cwd()) + r"\data\pcap" + r"\Processed Files\\"
dataPathProcessedCSV = str(Path.cwd()) + r"\data\csv" + r"\Processed Files\\"
```

```python
from ble import BleStream

stream = BleStream()
stream.from_pcap_file(dataPathInference + "Inference_V3.pcapng", parse_mode='tolerant', update=False,
                      ignore_error=True, fill_empty_packet=False)
```

```python
df = stream.to_pandas()
```

```python
df.to_parquet(dataPathProcessedCSV + "Inference_V3.parquet",
              engine="pyarrow",
              compression="snappy",
              index=False)
```

```python
from ble import BleStream

stream = BleStream()
stream.from_pcap_file(dataPathInference + "Inference_V2.pcapng", parse_mode='tolerant', update=False,
                      ignore_error=True, fill_empty_packet=False)

df = stream.to_pandas()

df.to_parquet(dataPathProcessedCSV + "Inference_V2.parquet",
              engine="pyarrow",
              compression="snappy",
              index=False)
```
