# Analysis of synthetically Generated Packets

```python
from pathlib import Path
import os

os.chdir(Path.cwd().parents[0])
print("Now in:", Path.cwd())

dataPathCSV = str(Path.cwd()) + r"\data\csv\\"
savePath = str(Path.cwd()) + r"\out\plots\synthetic_data_analysis\\"
```

```python
import pandas as pd
data = pd.read_csv(dataPathCSV + r"synthetic\synthetic_samples_other_device_train.csv", encoding="cp1252")
```

```python
from plotting.src.functions import plot_adv_type_distribution, plot_adv_count_distribution, plot_pdu_distribution, plot_hist_distribution

plot_adv_type_distribution(data, adv_column="AD Type", name="Distribution of Advertising Data Types", show=True, savePath=savePath)
plot_adv_count_distribution(data, adv_column="AD Type", name="Distribution of Number of Advertising Data Structs", show=True, savePath=savePath)
plot_pdu_distribution(data, pdu_column = "PDU", name = "Distribution of PDU Types", show = True, savePath=savePath)


plot_hist_distribution(data['Length Header'], x_label = "Length of Header in Bytes", name = "Probability Distribution of Header Length", show = True, savePath=savePath)

plot_hist_distribution( data['Service Data'].str.split(",").dropna().explode().str.len() / 2 , x_label = "Length of Service Data in Bytes", name = "Probability Distribution of Service Data Length", show = True, savePath=savePath)

plot_hist_distribution( data['MS Data'].str.split(",").dropna().explode().str.len() / 2, x_label = "Length of Manufacturer Specific Data in Bytes", name = "Probability Distribution of Manufacturer Specific Data Length", show = True, savePath=savePath)
```
